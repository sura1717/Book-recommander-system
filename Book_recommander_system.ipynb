```python
### Build a popularity based model(KNN)
```

```python
1. we display the top 50 books with highest rating with consider only those books which has got minimum 250 books
```

```python
# merge books and rating column on isbn
new_df = rating.merge(books,on="ISBN")
```

```python
new_df.head()
```

```python
num_df = new_df.groupby("Book-Title").count()["Book-Rating"].reset_index()
num_df.head()
```

```python
num_df.rename(columns = {"Book-Rating":"num-rating"},inplace=True)
num_df.head()
```

```python
avg_rating = new_df.groupby("Book-Title")["Book-Rating"].mean().reset_index()
avg_rating.head()
```

```python
avg_rating.rename(columns = {"Book-Rating":"Average-Rating"},inplace = True)
avg_rating
```

```python
# now merge both the dataframe 
populer_df = avg_rating.merge(num_df,on="Book-Title")
populer_df.head()
```

```python
# now filter top 50 movie with highest rating and more then 250 rating
populer_df = populer_df[populer_df["num-rating"]>=250].sort_values("Average-Rating",ascending=False)
populer_df.head()
```

```python
top_50 = populer_df.head(50)
top_50
```

```python
# top_50.merge(books,on="Book_Title") #.drop_duplicate("Book-Title")
top_50 = top_50.merge(books,on="Book-Title")
top_50.shape
```

```python
top_50 = top_50.drop_duplicates("Book-Title")
top_50.shape
```

```python
top_50.head()
```

```python
# now filter columns 
populer_50 = top_50[["Book-Title","Average-Rating","num-rating","Book-Author","Image-URL-M"]]
```

```python
populer_50
```

### colaborating filtering based recommander system

```python
# filter those user which is rate more then 200 books the chooice those book which is more then ratings 
```

```python
x = new_df.groupby("User-ID").count()["Book-Rating"]>200
```

```python
x_ind = x[x].index
```

```python
x_ind
```

```python
x_ind.shape
```

```python
filtered_rating = new_df[new_df["User-ID"].isin(x_ind)]
filtered_rating.shape
```

```python
# now filtered on the bases on books
y = filtered_rating.groupby("Book-Title").count()["Book-Rating"]>50
y
```

```python
famous_book = y[y].index
```

```python
final_rating = filtered_rating[filtered_rating["Book-Title"].isin(famous_book)]
final_rating
```

```python
# create a pivot table
pt = final_rating.pivot_table(index="Book-Title",columns = "User-ID",values = "Book-Rating")
```

```python
pt.fillna(0,inplace=True)
pt
```

### Now calculate the distances

```python
 # calculate eucledian distance
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
similarity_score[0]
```

```python
def recommand(book_name):
    index = np.where(pt.index == book_name)[0][0]
    similer_item = sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True)[1:6]

    data = []
    for i in similer_item:
        # print(pt.index[i[0]])
        item = []
        temp_df = books[books["Book-Title"]==pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates("Book-Title")["Book-Title"].values))
        item.extend(list(temp_df.drop_duplicates("Book-Title")["Book-Author"].values))
        item.extend(list(temp_df.drop_duplicates("Book-Title")["Image-URL-M"].values))


        data.append(item)
    return data
        

print(recommand("Zoya"))
```
